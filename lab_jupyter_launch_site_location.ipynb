Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# define a helper
def get_color(outcome):
    return 'green' if outcome == 1 else 'red'

# create the new column
spacex_df['marker_color'] = spacex_df['class'].apply(get_color)
spacex_df.tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# make sure marker_cluster is already added to site_map
site_map.add_child(marker_cluster)

# Loop through each launch record
for idx, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(
            color='white', 
            icon_color=row['marker_color'], 
            icon='rocket', 
            prefix='fa'
        ),
        popup=folium.Popup(
            f"{row['Launch Site']}<br><b>{'Success' if row['class']==1 else 'Failed'}</b>",
            parse_html=True
        )
    ).add_to(marker_cluster)

# Finally display the map
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# Example launch site (CCAFS LC-40) coords from launch_sites_df
launch_site = 'CCAFS LC-40'
site_lat, site_lon = launch_sites_df.loc[
    launch_sites_df['Launch Site']==launch_site, ['Lat','Long']
].values[0]

# Coordinates you read off as the nearest coastline
coastline_lat  = 28.5644  
coastline_lon  = -80.56845 

# Compute distance
distance_coastline = calculate_distance(
    site_lat, site_lon,
    coastline_lat, coastline_lon
)

print(f"Distance from {launch_site} to nearest coastline: {distance_coastline:.2f} km")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )

# your computed values
coordinate = [coastline_lat, coastline_lon]      # e.g. [28.56367, -80.57163]
distance = distance_coastline                    # in kilometers, float

# create and add the distance‐label marker
distance_marker = folium.Marker(
    location=coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=(
            '<div style="font-size:12pt; color:#d35400;">'
            f"<b>{distance:.2f} KM</b>"
            '</div>'
        )
    )
).add_to(site_map)

# Then display
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
from folium.features import DivIcon
import folium

# 1) Your base map and launch‐site coord
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
launch_site = 'CCAFS LC-40'
site_lat, site_lon = launch_sites_df.loc[
    launch_sites_df['Launch Site']==launch_site, ['Lat','Long']
].values[0]
site_coord = [site_lat, site_lon]

# 2) Define your proximity points (mouse‐picked coords)
proximities = [
    {'name':'Port', 'coord':[28.47513,  -80.78527]},
    {'name':'Nearest City', 'coord':[28.60415,  -80.82372 ]},
    {'name':'Railway',      'coord':[28.50913,  -80.78513 ]},
    {'name':'Highway',      'coord':[28.56155,  -80.57047 ]},
]

# 3) Loop, compute, and plot
for p in proximities:
    lat2, lon2 = p['coord']
    dist = calculate_distance(site_lat, site_lon, lat2, lon2)
    
    # a) distance marker
    folium.Marker(
        location=p['coord'],
        icon=DivIcon(
            icon_size=(120,30),
            icon_anchor=(0,0),
            html=(
                f'<div style="font-size:10pt;color:#d35400;">'
                f"<b>{p['name']}: {dist:.2f} km</b></div>"
            )
        )
    ).add_to(site_map)
    
    # b) connecting dashed line
    folium.PolyLine(
        locations=[site_coord, p['coord']],
        color='blue',
        weight=2,
        dash_array='5,5'
    ).add_to(site_map)

# 4) highlight the launch site itself
folium.Circle(
    location=site_coord,
    radius=500,
    color='black',
    fill=True,
    fill_opacity=0.2
).add_child(
    folium.Popup(launch_site)
).add_to(site_map)

# 5) display
site_map
```
